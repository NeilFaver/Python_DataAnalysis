```python
# seq = [7, 2, 3, 6, 3, 6, 0, 1]
seq[:5]
```

```python
# seq = [7, 2, 3, 6, 3, 6, 0, 1]
seq[3:]
```

*Negative indices* slice the sequence relative to the end

```python
seq[-4:]
```

```python
seq[-6:-2]
```

```python
seq[::2]
```

Check whether an object is hashable (can be used as a key in a dictionary) with the **hash** function

```python
hash("string") # The hash values you see when using the hash function in general will depend on the Python version you are using
```

```python
hash((1, 2, [2, 3])) # fails because lists are mutable
```

To use a list as a key, one option is to convert it to a tuple, which can be hashed as long as its elements also can be:

### <span style="color:violet">Set</span>

<div class="alert alert-block alert-info">
    A set is an <i>unordered</i> collection of <i>unique</i> elements.
</div>

A set can be created in two ways: via the *set function* or via a *set literal* with curly braces

```python
set([2, 2, 2, 1, 3, 3])
```

```python
{2, 2, 2, 1, 3, 3}
```

Sets support mathematical *set operations* like union, intersection, difference, and symmetric difference

```python
a = {1, 2, 3, 4, 5}
```

```python
b = {3, 4, 5, 6, 7, 8}
```

The union of these two sets is the set of distinct elements occurring in either set. This can be computed with either the union method or the | binary operator:

```python
a.union(b)
```

```python
a | b
```

The intersection contains the elements occurring in both sets. The & operator or the intersection method can be used:

```python
a.intersection(b)
```

```python
a & b
```

If you pass an input that is not a set to methods like union and intersection, Python will convert the input to a set before executing the operation. When using the binary operators, both objects must already be sets.

All of the logical set operations have in-place counterparts, which enable you to replace the contents of the set on the left side of the operation with the result. For very large sets, this may be more efficient:

```python
# a = {1, 2, 3, 4, 5}
c = a.copy()
```

```python
# b = {3, 4, 5, 6, 7, 8}
c |= b
```

```python
# b = {3, 4, 5, 6, 7, 8}
d &= b
```

```python
d
```

them to tuples:

```python
my_data = [1, 2, 3, 4]
```

```python
a_set = {1, 2, 3, 4, 5}
```

```python
# Check {1, 2, 3} is a subset of a_set
{1, 2, 3}.issubset(a_set)
```

```python
# Check a_set is a superset of {1, 2, 3}
a_set.issuperset({1, 2, 3})
```

Sets are equal if and only if their contents are equal:

```python
{1, 2, 3} == {3, 2, 1}
```

Itâ€™s common when iterating over a sequence to want to keep track of the index of the current item:

```python
collection = [1, 2, 3, 4]
index = 0
for value in collection:
# do something with value
    index += 1
index
```

Python has a built-in function, *enumerate*, which returns a sequence of (i, value) tuples

```python
collection = [1, 2, 3, 4]
for index, value in enumerate(collection):
# do something with value
    index += 1
index
```

```python
# enumerate function in loops
l1 = ["eat", "sleep", "repeat"]
```

```python
# printing the tuples in object directly
for ele in enumerate(l1):
    print (ele)
```

```python
# changing index and printing separately
for count, ele in enumerate(l1, 100):
    print (count, ele)
```

```python
# getting desired output from tuple
for count, ele in enumerate(l1):
    print(count)
    print(ele)
```

### <span style="color:violet">List, Set, and Dictionary Comprehensions</span>

*List comprehensions* are a convenient and widely used Python language feature. They allow you to concisely form a new list by filtering the elements of a collection, transforming the elements passing the filter into one concise expression. They take the basic form:
<br>[*expr* for value in collection if *condition*]</br>

This is equivalent to the following for loop:
<br>result = []</br>
for value in collections:
> if *condition*:
> > result.append(*expr*)

The filter condition can be omitted, leaving only the expression

```python
strings = ["a", "as", "bat", "car", "dove", "python"]
```

```python
[x.upper() for x in strings if len(x) > 2]
```

Set and dictionary comprehensions are a natural extension, producing sets and dictionaries in an idiomatically similar way instead of lists.
